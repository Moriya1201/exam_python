## 4.3.3 グラフの種類と出力方法

### 折れ線グラフ

```python
# 描画オブジェクト(fig)とサブプロット(ax)を生成
fig, ax = plt.subplots()

x = [1, 2, 3]
y1 = [1, 2, 3]
y2 = [3, 1, 2]

ax.plot(x, y1) # 折れ線グラフを描画
ax.plot(x, y2)

plt.show()
```

```python
# sin cos のグラフ
import numpy as np

x =np.arange(0.0, 15.0, 0.1)
y1 = np.sin(x)
y2 = np.cos(x)

fig, ax = plt.subplots()

ax.plot(x, y1, label = 'sin') # 折れ線グラフを描画
ax.plot(x, y2, label = 'cos')
ax.legend(loc = 'best')

plt.show()
```

### 棒グラフ

```python
fig, ax = plt.subplots()

x = [1, 2, 3]
y = [10, 2, 3]
ax.bar(x, y) # 棒グラフを描画

plt.show()
```

```python
# barメソッドのtick_label引数を使用するとメモリに任意のラベルが指定できる

fig, ax = plt.subplots()

x = [1, 2, 3]
y = [10, 2, 3]

labels = ['spam', 'ham', 'egg']
ax.bar(x, y, tick_label = labels) # 棒グラフを描画

plt.show()
```

```python
# 横向きの棒グラフ　bar→barh

fig, ax = plt.subplots()

x = [1, 2, 3]
y = [10, 2, 3]

labels = ['spam', 'ham', 'egg']
ax.barh(x, y, tick_label = labels) # 棒グラフを描画

plt.show()
```

```python
# 複数の棒グラフを並べて表示する場合→棒グラフの幅を指定してずらして描画する

fig, ax = plt.subplots()

x = [1, 2, 3]
y1 = [10, 2, 3]
y2 = [5, 3, 6]
labels = ['spam', 'ham', 'egg']

width = 0.3 #棒グラフの幅を0.4にする
ax.bar(x, y1, width = width, tick_label = labels, label = 'y1') # 幅を指定して描画

# 幅分ずらして棒グラフを描画
x2 = [num + width for num in x] #この意味？？？
ax.bar(x2, y2, width = width, label = 'y2')

ax.legend()

plt.show()
```

```python
# 積み上げ棒グラフ：両方を足した棒グラフを描画し、次に片方の値のみを棒グラフを描画する　
# 積み上げる数が増えてもやることは同じ

fig, ax = plt.subplots()

x = [1, 2, 3]
y1 = [10, 2, 3]
y2 = [5, 3, 6]
labels = ['spam', 'ham', 'egg']

# y1とy２を足した値を格納
y_total = [num1 + num2 for num1, num2 in zip(y1, y2)]

# y1とy2を足した高さの棒グラフを描画
ax.bar(x, y_total, tick_label = labels, label = 'y1')
ax.bar(x, y2, label = 'y2')
ax.legend()

plt.show()
```

### 散布図

```python
# ランダムに生成した５０個の要素を散布図で描画

fig, ax = plt.subplots()

# ランダムに50個の要素を生成
np.random.seed(123)
x = np.random.rand(50)
y = np.random.rand(50)

ax.scatter(x, y) #散布図を描画

plt.show()
```

```python
# marker引数にマーカーの形を指定することができる
# 10こずつ異なる形のマーカーを使用

fig, ax = plt.subplots()

# ランダムに50個の要素を生成
np.random.seed(123)
x = np.random.rand(50)
y = np.random.rand(50)

ax.scatter(x[0:10], y[0:10], marker = 'v', label = 'triangle down') # 下向き三角
ax.scatter(x[10:20], y[10:20], marker = '^', label = 'triangle up') # 上向き三角
ax.scatter(x[20:30], y[20:30], marker = 's', label = 'square') # 正方形
ax.scatter(x[30:40], y[30:40], marker = '*', label = 'star') # 星形
ax.scatter(x[40:50], y[40:50], marker = 'x', label = 'x') # X

ax.legend()

plt.show()



```

### ヒストグラム

```python
# データを生成
np.random.seed(123)
mu = 100 #平均値
sigma = 15 #標準偏差
x = np.random.normal(mu, sigma, 1000)

fig, ax = plt.subplots()

# ヒストグラムを描画
n, bins, patches = ax.hist(x) 

plt.show()
```

```python
# n: 各bin（棒）の度数、bins:ビンの境界の値、patches:ビンを描画するための情報

# nとbinを使用すると以下のコードで度数分布表が出力できる
for i, num in enumerate(n):
    print('{:.2f} - {:.2f}: {}'.format(bins[i], bins[i + 1], num))
```

```python
# histメソッドのbins引数に任意の数値を指定するとビン（棒）の数を変更できる （デフォルトは10）

fig, ax = plt.subplots()
ax.hist(x, bins =25) # ビンの数を指定して描画

plt.show()
```

```python
# histメソッドの引数に orientation = 'horizontal'と指定すると横向きのヒストグラムを描画

fig, ax = plt.subplots()
ax.hist(x, orientation = 'horizontal') # ビンの数を指定して描画

plt.show()
```

```python
# ヒストグラムでは複数の値を指定すると自動的に横に並べて表示してくれる（⇄棒グラフの場合指定が必要）

np.random.seed(123)
mu = 100 #平均値

x0 = np.random.normal(mu, 20, 1000)

#異なる標準偏差でデータを作成
x1 = np.random.normal(mu, 15, 1000)
x2 = np.random.normal(mu, 10, 1000)

fig, ax = plt.subplots()

labels = ['x0', 'x1', 'x2']

# 3つのデータのヒストグラムを描画
ax.hist((x0, x1, x2), label = labels) 
ax.legend()

plt.show()
```

```python
# 積み上げヒストグラム：histメソッドの引数に　stacked = True と指定

fig, ax = plt.subplots()

labels = ['x0', 'x1', 'x2']

# 積み上げたヒストグラムを描画
ax.hist((x0, x1, x2), label = labels, stacked = True) 
ax.legend()

plt.show()
```

### 箱ひげ図

```python
# boxplotメソッドを使用

#異なる標準偏差でデータを作成

np.random.seed(123)
x0 = np.random.normal(mu, 20, 1000)
x1 = np.random.normal(mu, 15, 1000)
x2 = np.random.normal(mu, 10, 1000)

fig, ax = plt.subplots()
labels = ['x0', 'x1', 'x2']

# 3つのデータの箱ひげ図を描画
ax.boxplot((x0, x1, x2), labels = labels) # labels = labels と左辺にもsがついている　？？？

plt.show()
```

```python
# boxplotの引数に　vert = False と指定すると横向きの箱ひげ図を描画

fig, ax = plt.subplots()
labels = ['x0', 'x1', 'x2']

# 横向きの箱ひげ図を描画
ax.boxplot((x0, x1, x2), labels = labels, vert = False) 

plt.show()
```

### 円グラフ

```python
# pieメソッドで描画する
```

```python
labels = ['spam', 'ham', 'egg']
x = [10, 3 ,1]

fig, ax = plt.subplots()

ax.pie(x, labels = labels) # 円グラフを描画

plt.show()
```

```python
# 描画領域に合わせて楕円に変化した場合の対応方法　アスペクト比を保持するためにaxis('equal')と指定

fig, ax = plt.subplots()

ax.pie(x, labels = labels) # 円グラフを描画
ax.axis('equal')

plt.show()
```

```python
# デフォルトでは15時の1から反時計回りに要素が配置
# 12時から描画を始める→startangle = 90(90度の位置)
#  時計回りに配置→counterclock = False （反時計回りを否定する）

fig, ax = plt.subplots()

ax.pie(x, labels = labels, startangle = 90, counterclock = False) # 上から時計回り
ax.axis('equal')

plt.show()
```

```python
#　円グラフに影をつけ、値のパーセント表記を追記

fig, ax = plt.subplots()

ax.pie(x, labels = labels, startangle = 90, counterclock = False, shadow = True, autopct = '%1.2f%%') #影とパーセント表記を追加
ax.axis('equal')

plt.show()
```

```python
# 一部の値を強調するための切り出し表示：explode

explode = [0, 0.2, 0] #1番目の要素(ham)を切り出す

fig, ax = plt.subplots()

ax.pie(x, labels = labels, startangle = 90, counterclock = False, 
       shadow = True, autopct = '%1.2f%%',
      explode = explode) #explodeを指定
ax.axis('equal')

plt.show()
```

### 複数のグラフを組み合わせる

```python
# 棒グラフと折れ線グラフを1つのサブプロット上に描画

fig, ax = plt.subplots()

x1 = [1, 2, 3]
y1 = [5, 2, 3]
x2 = [1, 2, 3, 4]
y2 = [8, 5, 4, 6]

ax.bar(x1, y1, label = 'y1') # 棒グラフを描画
ax.plot(x2, y2, label = 'y2') # 折れ線グラフを描画
ax.legend()

plt.show()
```

```python
# ヒストグラムに折れ線グラフを追加して近似曲線の描画

np.random.seed(123)
x = np.random.randn(1000) #正規乱数を生成

fig, ax = plt.subplots()

# ヒストグラムを描画
counts, edges, patches = ax.hist(x, bins = 25)

# 近似曲線に用いる点を求める（ヒストグラムのビンの中点）
x_fit = (edges[:-1] + edges[1:])/2

# 近似曲線をプロット
y = 1000 * np.diff(edges) * np.exp(-x_fit**2 / 2) / np.sqrt(2 * np.pi) # ???
ax.plot(x_fit, y)

plt.show()
```

### フォント

```python
# タイトル、凡例、軸ラベルなどのテキストに対してステイルが設定できる
# size引数 : フォントサイズ
# weight引数 : フォントの太さ（light,　bold）
# family引数 : フォントの種類　デフォルトで使用できるのはserif, sans-serif, cursive, fantasy, monospace
```

```python
fig, ax = plt.subplots()

ax.set_xlabel('xlabel', family='fantasy', size=20,
              weight='bold')
ax.set_ylabel('ylabel', family='cursive', size=40,
              weight='light')
ax.set_title('graph title', family='monospace',
             size=25, weight='heavy')

plt.show()
```

```python
# 同じフォントの指定を様々な箇所で使う場合→フォントの設定を辞書データとして作成、fontdict引数に一度に指定
```

```python
# フォントのスタイルを辞書で定義
fontdict = {
    'family': 'fantasy',
    'size': 20,
    'weight': 'normal',
}

fig, ax = plt.subplots()

# 辞書形式でフォントのスタイルを指定
ax.set_xlabel('xlabel', fontdict=fontdict)
ax.set_ylabel('ylabel', fontdict=fontdict)
# 個別指定したsizeで上書き可能
ax.set_title('graph title', fontdict=fontdict, size=40)

plt.show()
```

### pandasのオブジェクトからグラフ描画

```python
# pandasのdf、Seriesからグラフを描画できる（内部的にはMatplotlibを使用）
# 複雑な組み合わせなどはできないが、簡単な可視化に向いている
```

```python
import pandas as pd
import matplotlib.style
import matplotlib.pyplot as plt

matplotlib.style.use('ggplot')  # スタイルを指定

# DataFrameを作成
df = pd.DataFrame({'A': [1, 2, 3], 'B': [3, 1, 2]})
df.plot()  # DataFrameに対してplotメソッドを呼び出し、折れ線グラフを描画
plt.show()
```
